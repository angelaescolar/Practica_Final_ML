Vamos a ver la correlaciÃ³n entre estas variables numÃ©ricas ya que posteriormente puede mejorar nuestro estudio eliminar aquellas que aporten gran cantidad de informaciÃ³n similar.
TambiÃ©n vamos a ver aquellas que tengan una correlaciÃ³n por encima de 0,6

```python
df[Var_continuas].corr(method='pearson')
```

```python
corr = df[Var_continuas].corr('pearson')
new_corr = corr.abs()
new_corr.loc[:,:] = np.tril(new_corr, k=-1)
new_corr = new_corr.stack().to_frame('correlation').reset_index().sort_values(by='correlation', ascending=False)
new_corr[new_corr['correlation']>0.6]
```
